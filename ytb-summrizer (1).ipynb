```python
def normalize_text(text):
    text = text.replace("\xa0", " ")
    return " ".join(text.split())


def extract_video_id(url: str) -> str:
    parsed = urlparse(url)

    qs = parse_qs(parsed.query)
    if 'v' in qs:
        return qs['v'][0]

    if 'youtu.be' in parsed.netloc:
        return parsed.path.strip('/')

    path_parts = parsed.path.strip('/').split('/')
    if len(path_parts) >= 2 and path_parts[0] in {'shorts', 'embed'}:
        return path_parts[1]

    raise ValueError("Invalid YouTube URL")


def get_youtube_transcript(url: str) -> str:
    video_id = extract_video_id(url)
    fetched = YouTubeTranscriptApi().fetch(video_id, languages=['en'])
    text = " ".join(snippet.text for snippet in fetched)
    return normalize_text(text)


def chunk_text(text, tokenizer, n=900):
    tokens = tokenizer.encode(text)
    for i in range(0, len(tokens), n):
        yield tokenizer.decode(tokens[i:i+n])


def summary_chunks(text):
    summaries = []
    for chunk in chunk_text(text, summarizer.tokenizer):
        if len(chunk.split()) < 150:
            continue
        s = summarizer(chunk, max_length=300, min_length=150, do_sample=False)
        summaries.append(s[0]["summary_text"])
    return summaries


def hierarchical_summary(text):
    first_pass = summary_chunks(text)
    combined = " ".join(first_pass)

    final = summarizer(
        combined,
        max_length=300,
        min_length=150,
        do_sample=False
    )

    return final[0]["summary_text"]
```

```python
from fastapi import FastAPI, Request, HTTPException
import uvicorn
import threading
import time
import socket
from urllib.parse import urlparse, parse_qs
from pyngrok import ngrok, conf
app = FastAPI()
@app.post("/summarize")
async def summarize(req: Request):
    if req.headers.get("authorization") != f"Bearer {API_KEY}":
        raise HTTPException(status_code=401, detail="Unauthorized")

    data = await req.json()
    input_type = data.get("type")
    content = data.get("content")

    if not input_type or not content:
        raise HTTPException(status_code=400, detail="Invalid request")

    if input_type == "youtube":
        text = get_youtube_transcript(content)
    elif input_type == "text":
        text = normalize_text(content)
    else:
        raise HTTPException(status_code=400, detail="Unknown type")

    summary = hierarchical_summary(text)

    return {
        "summary": summary
    }

# ---------------- NGROK ----------------
def free_port():
    s = socket.socket()
    s.bind(('', 0))
    port = s.getsockname()[1]
    s.close()
    return port

port = free_port()
conf.get_default().auth_token = NGROK_TOKEN
public_url = ngrok.connect(port).public_url
print("Public URL:", public_url)

def run():
    uvicorn.run(app, host="0.0.0.0", port=port)

threading.Thread(target=run, daemon=True).start()
time.sleep(1)

```
